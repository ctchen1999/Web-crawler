```python
# Import packages
import requests
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.select import Select
from bs4 import BeautifulSoup
import time

import pandas as pd
```

```python
# Initialization
save_dict = {"title": [],
             "date": [],
             "author": [],
             }

options = Options()
# options.add_argument("--headless")
```

```python
driver = webdriver.Chrome(options=options)

# timer = threading.Timer(5, driver.quit(), args=[driver])

# 進入高鐵首頁
driver.get("https://www.ptt.cc/bbs/music/index.html")
driver.implicitly_wait(3) # Loading wait⋯
TOTAL_PAGES = int(driver.find_elements(By.CLASS_NAME, "btn")[-3].get_attribute("href").split("index")[-1].split(".")[0])
print("Total pates to be crawled: ", TOTAL_PAGES)

for idx in range(TOTAL_PAGES):
    start_time = time.time()
    
    
    soup = BeautifulSoup(driver.page_source)
    
    blocks = soup.find_all("div", class_="r-ent")
    for block in blocks:
        article = block.find("div", class_="title")
        save_dict["title"].append(article.a.text)
        date    = block.find("div", class_="date")
        save_dict["date"].append(date.text)
        author  = block.find("div", class_="author")
        save_dict["author"].append(author.text)
    
    # 上一頁
    driver.find_elements(By.CLASS_NAME, "btn")[-3].click()
    print("PAGE:", idx)
    
    end_time = time.time()
    process_time = end_time - start_time

    # timer.cancel()
# time.sleep(5)
# driver.close()
```
